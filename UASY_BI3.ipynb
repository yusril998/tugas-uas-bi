```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
caesarian_df = pd.read_excel("D:/UASY_BI/data_UAS-main/caesarian.xlsx",
                             delimiter=',',
                       usecols=['Usia','Kelahiran_ke-','Waktu_Kelahiran','Tekanan_darah','Kelainan_jantung','Caesarian'],
                       dtype={'Usia': 'int32','Kelahiran_ke-': 'int32','Waktu_Kelahiran': 'int32','Tekanan_darah': 'int32','Kelainan_jantung': 'int32',
                              'Caesarian': 'int32'})
```

```python
caesarian_df.head()
```

```python
caesarian_df.shape
```

# 3A

```python
import math
dis = []
for i in range(80):
    dis.append(math.sqrt((float(caesarian_df.iloc[i]['Usia'])-25)**2+ 
                         (float(caesarian_df.iloc[i]['Kelahiran_ke-'])-1)**2+
                         (float(caesarian_df.iloc[i]['Waktu_Kelahiran'])-0)**2+
                         (float(caesarian_df.iloc[i]['Tekanan_darah'])-1)**2+ 
                         (float(caesarian_df.iloc[i]['Kelainan_jantung'])-0)**2))
```

```python
caesarian_df['dis'] = dis
caesarian_df
```

```python
y=caesarian_df.sort_values('dis').head(5)
y
```

```python
z=y["Caesarian"]
z
```

```python
np.mean(z)
```

Kesimpulan,ditemukan hasil bahwa ibu hamil dengan kondisi tersebut diberlakukan 0,4 atau jika dibulatkan menjadi 0 maka akan bersalin secara normal
